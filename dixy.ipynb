```python
import pandas as pd
```

```python
df = pd.read_excel('data-28509-2018-10-13.xlsx')
```

```python
df.head()
```

```python
len(df)
```

```python
df_dixy = pd.DataFrame(columns= ['latitude', 'longitude', 'address'])
```

```python
df_perekrestok = pd.DataFrame(columns= ['latitude', 'longitude', 'address'])
```

```python

df_magnit = pd.DataFrame(columns= ['latitude', 'longitude', 'address'])
```

```python
df_billa = pd.DataFrame(columns= ['latitude', 'longitude', 'address'])
```

```python
for i in range(len(df)):
    s = df.Name[i]
    t = s.lower()
    l = df.geoData[i]
    if t.find('дикси') > -1:
        df_dixy.set_value(len(df_dixy), 'address', df.Address[i])
        df_dixy.longitude[len(df_dixy)-1] = l[l.find('[')+1:l.rfind(',')]
        df_dixy.latitude[len(df_dixy)-1] = l[l.rfind(',')+2:l.find(']')]
        print(df_dixy.longitude[len(df_dixy)-1])
    if t.find('билла') > -1:
        df_billa.set_value(len(df_billa), 'address', df.Address[i])
        df_billa.longitude[len(df_billa)-1] = l[l.find('[')+1:l.rfind(',')]
        df_billa.latitude[len(df_billa)-1] = l[l.rfind(',')+2:l.find(']')]
        
    if t.find('перекресток') > -1:
        df_perekrestok.set_value(len(df_perekrestok), 'address', df.Address[i])
        df_perekrestok.longitude[len(df_perekrestok)-1] = l[l.find('[')+1:l.rfind(',')]
        df_perekrestok.latitude[len(df_perekrestok)-1] = l[l.rfind(',')+2:l.find(']')]
        
    if t.find('магнит') > -1:
        df_magnit.set_value(len(df_magnit), 'address', df.Address[i])
        df_magnit.longitude[len(df_magnit)-1] = l[l.find('[')+1:l.rfind(',')]
        df_magnit.latitude[len(df_magnit)-1] = l[l.rfind(',')+2:l.find(']')]
        
    
    
    
```
