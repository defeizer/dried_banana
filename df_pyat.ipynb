```python
from requests import get
import re
 
r = get('https://5ka.ru/api/stores/?bbox=10.0000,10.0000,100.0000,100.0000')
text_stores = r.text
 
coordinates = re.findall(r'"coordinates":(.*?)},"properties"', text_stores)
 
print(len(coordinates))
 
print(coordinates)
```

```python
import pandas as pd
```

```python
df_pyat = pd.DataFrame({
    'latitude' : coordinates,
    'longitude' :  coordinates,
    'address' : ''
        })
```

```python
for i in range(len(df_pyat)):
    df_pyat.latitude[i] = df_pyat.latitude[i][1:df_pyat.latitude[i].find( ',')]
    df_pyat.longitude[i] = df_pyat.longitude[i][df_pyat.longitude[i].find( ',')+1:-1]
```
